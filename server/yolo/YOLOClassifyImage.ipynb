```python
import os
import random
import shutil

# Define paths
original_root = '/content/drive/MyDrive/first-half'
destination_root = '/content/YOLOClassify'
train_root = os.path.join(destination_root, 'train')
valid_root = os.path.join(destination_root, 'val')

# Create train and valid directories if they don't exist
os.makedirs(train_root, exist_ok=True)
os.makedirs(valid_root, exist_ok=True)

# Function to split dataset
def split_dataset(source_folder, train_folder, valid_folder, split_ratio=0.8):
    # Iterate through each folder (indoor, outdoor, retail store)
    for folder_name in os.listdir(source_folder):
        folder_path = os.path.join(source_folder, folder_name)
        if os.path.isdir(folder_path):
            # Create corresponding folders in train and valid directories
            train_folder_path = os.path.join(train_folder, folder_name)
            valid_folder_path = os.path.join(valid_folder, folder_name)
            os.makedirs(train_folder_path, exist_ok=True)
            os.makedirs(valid_folder_path, exist_ok=True)

            # List images in the current folder
            images = [f for f in os.listdir(folder_path) if f.endswith('.jpg') or f.endswith('.png')]

            # Shuffle images randomly
            random.shuffle(images)

            # Split images based on the split_ratio
            split_index = int(len(images) * split_ratio)
            train_images = images[:split_index]
            valid_images = images[split_index:]

            # Move images to train folder
            for image in train_images:
                src = os.path.join(folder_path, image)
                dest = os.path.join(train_folder_path, image)
                shutil.copy(src, dest)

            # Move images to valid folder
            for image in valid_images:
                src = os.path.join(folder_path, image)
                dest = os.path.join(valid_folder_path, image)
                shutil.copy(src, dest)

# Call function to split dataset
split_dataset(original_root, train_root, valid_root)

# Output confirmation
print(f"Dataset split into train and valid sets in '{train_root}' and '{valid_root}' respectively.")
```
