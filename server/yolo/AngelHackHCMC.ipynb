## Create working folders
- Creating images in train and validating sets
(First-time use only)

```python
import shutil
import os

root = "/content"
image_path = "/content/drive/MyDrive/FULL [Heineken Vietnam] Developer Resources"
label_path = "/content/drive/MyDrive/YOLO_Heineken_Label/obj_train_data"

for subdir in ['train', 'valid']:
    os.makedirs(os.path.join(root, f"labels/{subdir}"), exist_ok = True)
    os.makedirs(os.path.join(root, f"images/{subdir}"), exist_ok = True)

labels = [f for f in os.listdir(label_path) if f.endswith('.txt')]

train_ratio = 0.8
valid_ratio = 0.2

num_train = int(len(labels) * train_ratio)

train_labels = labels[:num_train]
valid_labels = labels[num_train:]

print(len(train_labels))
print(len(valid_labels))

for label in train_labels:
    shutil.copy(os.path.join(image_path, label.replace('.txt', '.jpg')),"/content/images/train")
    shutil.copy(os.path.join(label_path, label), "/content/labels/train")

for label in valid_labels:
    shutil.copy(os.path.join(image_path, label.replace('.txt', '.jpg')),"/content/images/valid")
    shutil.copy(os.path.join(label_path, label), "/content/labels/valid")

name_path = "/content/drive/MyDrive/YOLO_Heineken_Label/obj.names"
config_path = "/content/build.yaml"

with open(name_path, 'r') as file:
    class_names = file.read().strip().split('\n')

config_content = """names:
"""
for index, name in enumerate(class_names):
    config_content += f"  {index}: {name}\n"

config_content += "nc: 36\n" # Number of classes here
config_content += """train: /content/images/train
val: /content/images/valid
"""

with open(config_path, "w") as file:
    file.write(config_content)

print(config_content)
```

# PREDICT PHASE

```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolov8n.pt")  # pretrained YOLOv8n model

# Run batched inference on a list of images
results = model(["im1.jpg", "im2.jpg"])  # return a list of Results objects

# Process results list
for result in results:
    boxes = result.boxes  # Boxes object for bounding box outputs
    masks = result.masks  # Masks object for segmentation masks outputs
    keypoints = result.keypoints  # Keypoints object for pose outputs
    probs = result.probs  # Probs object for classification outputs
    obb = result.obb  # Oriented boxes object for OBB outputs
    result.show()  # display to screen
    result.save(filename="result.jpg")  # save to disk
```
